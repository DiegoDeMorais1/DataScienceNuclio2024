# Logistic Regression
In this exercise, we will use the Iris dataset to apply logistic regression to predict if a flower belongs to a specific species considering features like petal length and width.

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
from sklearn import datasets
iris = datasets.load_iris()
```

## Inspect the dataset





```
print(list(iris.keys()))
```

```
print(iris.DESCR)
```

## Select dependent and independent variables

```
X = iris["data"][:, 3:]  # petal width
y = (iris["target"] == 2).astype(int)  # 1 if Iris virginica, else 0
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Logistic Regression model on the Training set

```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 42)
classifier.fit(X_train, y_train)
```

## Predicting a new result

```
print(classifier.predict([[1.5]]))
```

## Predicting the Test set results

```
y_pred = classifier.predict(X_test)
```

## Evaluating the model with R2, MAE, MSE and RMSE

```
from sklearn.metrics import accuracy_score, f1_score, average_precision_score, recall_score

test_accuracy = accuracy_score(y_test, y_pred)
test_f1 = f1_score(y_test, y_pred)
test_precision = average_precision_score(y_test, y_pred)
test_recall = recall_score(y_test, y_pred)

print("Accuracy: {}".format(test_accuracy))
print("F1: {}".format(test_f1))
print("Precision: {}".format(test_precision))
print("Recall: {}".format(test_recall))
```

## Making and displaying the Confusion Matrix

```
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=classifier.classes_)

disp.plot()
plt.show()
```

## Estimated probabilities for flowers with petal widths from 0 to 3 cm

```
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = classifier.predict_proba(X_new)

plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris virginica")
```

## Estimated probabilities and decision boundaries

```
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = classifier.predict_proba(X_new)
decision_boundary = X_new[y_proba[:, 1] >= 0.5][0]

plt.figure(figsize=(8, 3))
plt.plot(X[y==0], y[y==0], "bs")
plt.plot(X[y==1], y[y==1], "g^")
plt.plot([decision_boundary, decision_boundary], [-1, 2], "k:", linewidth=2)

plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris virginica")

plt.xlabel("Petal width (cm)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 3, -0.02, 1.02])

plt.show()
```

## Select new features (independent variables) from the dataset


> We will use petal length and petal width




```
X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"] == 2).astype(int)
```

## Retrain the logistic regressor

```
classifier = LogisticRegression(solver="lbfgs", C=10**10, random_state=42)
classifier.fit(X, y)
```

## Inspect the linear decision boundary

```
x0, x1 = np.meshgrid(
        np.linspace(2.9, 7, 500).reshape(-1, 1),
        np.linspace(0.8, 2.7, 200).reshape(-1, 1),
    )

X_new = np.c_[x0.ravel(), x1.ravel()]

y_proba = classifier.predict_proba(X_new)

plt.figure(figsize=(10, 4))
plt.plot(X[y==0, 0], X[y==0, 1], "bs")
plt.plot(X[y==1, 0], X[y==1, 1], "g^")

# linear decision boundary
left_right = np.array([2.9, 7])
boundary = -(classifier.coef_[0][0] * left_right + classifier.intercept_[0]) / classifier.coef_[0][1]
plt.plot(left_right, boundary, "k--", linewidth=3)

zz = y_proba[:, 1].reshape(x0.shape)
#https://matplotlib.org/3.5.0/tutorials/colors/colormaps.html

# parallel lines with different probability values
contour = plt.contour(x0, x1, zz, cmap=plt.cm.brg)
plt.clabel(contour, inline=1, fontsize=12)

# colour background for areas/regions
custom_cmap = ListedColormap(['#fafab0','#a0faa0'])
plt.contourf(x0, x1, zz, cmap=custom_cmap)

plt.text(3.5, 1.5, "Not Iris virginica", fontsize=14, color="b", ha="center")
plt.text(6.5, 2.3, "Iris virginica", fontsize=14, color="g", ha="center")
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.axis([2.9, 7, 0.8, 2.7])
plt.show()
```
