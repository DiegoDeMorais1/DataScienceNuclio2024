```python
# Importar as bibliotecas necessrias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Gerar dados de exemplo
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Instanciar o modelo KMeans com 4 clusters
kmeans = KMeans(n_clusters=4)

# Ajustar o modelo aos dados
kmeans.fit(X)

# Obter os centrides dos clusters e os rtulos dos clusters para cada amostra
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Plotar os dados e os clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='x')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KMeans Clustering')
plt.show()
```
